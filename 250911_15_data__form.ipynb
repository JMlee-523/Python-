### 제곱
* x의 y제곱은 ** 연산자를 활용한다.

* 예를 들어, (10 * 2 + 1)^2 를 파이썬을 통해 연산하고자 한다면

### 몫과 나머지

* `a % b`와 같이 코드를 작성하면 a를 b로 나눈 나머지를 계산한다.

* `a // b` 와 같이 코드를 작성하면 a를 b로 나눈 몫를 계산한다.

* //, % 연산자 > a, b가 정수형이면 결과도 정수형

* / 연산자 > a,b값에 관계 없이 실수형

* 따라서 `a = b*(a//b)+a%b ` 는 항상 성립하나 `a = b*(a/b)` 는 성립하지 않을 수 있다

```
'현재 온도는 18도입니다'
```
%nbsp; 와 같이 출력해주는 프로그램을 작성했다고 가정하자. 온도가 20도로 변화한다면
```
'현재 온도는 20도입니다'
```
&nbsp; 와 같은 문장이 출력될 것이다. 온도를 제외한 문자열은 모두 같은데 20과 18이라는 숫자만 다르다. 이와 같이 문자열 내의 특정한 값만을 바꿔야 할 경우 사용하는 기능이 포매팅이다.

 ```
  " I raised %d UMAs. " %3
  ' I raised 3 UMAs '
  ```

  &nbsp; 위의 '%d' 는 문자열 포맷 코드라고 부르며, 해당 코드는 정수를 위한 코드이다.

  ```
  " I raised %s UMAs. " %five
  ' I raised five UMAs.'
  ```

  &nbsp; '%s' 는 문자열을 삽입하기 위한 문자열 포맷 코드이다.

 |  코드 |   용도  |         어원         |
| :-: | :---: | :----------------: |
|  %s |  문자열  |     **s**tring     |
|  %c | 문자 1개 |    **c**haracter   |
|  %d |   정수  | **d**ecimal (10진수) |
|  %f |  부동소수 | **f**loating point |
|  %o |  8진수  |      **o**ctal     |
|  %x |  16진수 |   he**x**adecimal  |
|  %% |  문자 % |       % 자체 출력      |
<br>

### 포맷 코드 활용하기

 1. 정렬과 공백
  
  ```
  "%10s" %"hi"
  '        hi'
  ```

  %10s는 전체 길이가 10개인 문자열 공간에서 대입되는 값을 오른쪽으로 정렬하고 그 나머지는 공백으로 남겨 두라는 의미이다.

  반대로, 왼쪽 정렬을 원한다면 - 연산자를 사용하여 "%-10s"와 같이 사용할 수 있다. 총 길이 10개 중 왼쪽부터 정렬하고, 나머지는 남겨 두라는 의미.

  ```
  "%-10s" %"hi"
  'hi        '
  ```
  이때, 지정한 문자열의 길이는 **최소 길이**이다. 즉, 문자열의 길이가 10개 이상이라면 문자열의 길이에 맞게 출력도 길어진다.
  ```
  a='neo universe' # len(a) = 12
  "%-10s" %a
  'neo universe'
  ```
  2. 소수점 표현하기
  ```
  "%0.4f" %3.1415926535
  '3.1416'
  ```
  원주율을 소수점 네 번째 자리까지 나타내고 싶다면 위와 같이 작성할 수 있다. %0.4f에서 '.'은 소수점 포인트, 그 뒤의 숫자 4는 소수점 뒤에 나올 숫자의 개수를 말한다. 소수점 포인트 앞의 숫자는 문자열의 전체 길이를 의미한다. %0.4f에서 사용한 숫자 0은 길이에 상관하지 않겠다는 의미이며, 생략하여 %.4f와 같이 나타낼 수도 있다.

  ### format 함수를 사용한 포매팅

  format 함수를 사용하면 더 발전된 스타일로 포맷을 지정할 수 있다. 앞에서 살펴본 문자열 포매팅 예제를 format 함수를 사용하여 바꾸면 다음과 같다.

  * 숫자 바로 대입하기
  ```
  "I raised {0} UMAs" .format(3)
 ```

&nbsp; 문자열 중 {0} 부분이 .format()의 괄호 안에 입력한 3으로 바뀐 것을 확인할 수 있다.

<br>

 * 문자열 바로 대입하기
 ```
 "I raised {0} UMAs" .format("five")
 ```

 &nbsp; 문자열의 {0} 항목이 'five'라는 문자열로 바뀌었다.

 <br>

  * 둘 이상의 값 넣기

  ```
  number = 2
  time = 60
  " I raised {0} UMAs. so I spent {1} minutes." .format(number,time)
 >> 'I raised 2 UMAs. so I spent 60 minutes.'
 ```
 &nbsp; 2개 이상의 값을 넣을 경우, 문자열의 {0}, {1}과 같은 인덱스 항목이 format 함수의 입력값으로 순서에 맞게 바뀐다.

 <br>
  
  * 이름으로 넣기
  ```
  " I raised {number} UMAs. so I spent {time} minutes." .format(number=2,time=60)
  ```
  &nbsp; {0}, {1}과 같은 형태보다 더 편리한 {name} 형태 또한 사용할 수 있다. 이 경우 format 함수 내에 name=value 와 같은 꼴의 입력값이 있어야 한다. 인덱스와 이름을 혼용하는 것 또한 가능하다.

 <br>

 * 좌우 정렬

 ```
 "{0:<10}".format("hi")
 >> "hi        "
 ```

&nbsp; :<10 표현식을 사용하면 치환되는 문자열을 왼쪽으로 정렬하고 총 자릿수를 10으로, :>10을 사용하면 문자열을 오른쪽으로 정렬하고 총 자릿수를 10으로 맞출 수 있다. <br>
&nbsp; 같은 원리로 :^10 표현식을 사용하면 가운데 정렬 또한 가능하다.

 <br>

  * 공백 채우기
  
  ```
  "{0:=^10"}".format{"hi"}
  '====hi===='
  ```
 &nbsp;  위와 같이, 채워넣고자 하는 문자를 정렬 문자 <,>,^ 앞에 입력함으로서 공백 대신 문자 값으로 채워 넣는 것 또한 가능하다. <br>

  * 소수점 표현하기

  ```
  y=3.141592
  "{0:0.4f}".format(y)
  >>> '3.1416'
  ```
 &nbsp; 앞에서 사용했던 .4f와 같은 표현식을 그대로 사용하여 소숫점을 표현할 수 있다. <br><br>

  ### f 문자열 포매팅

 &nbsp; 최신 파이썬 기준, 가장 널리 사용되는 포매팅 방법은 f 문자열 포매팅이다. 다음과 같이, 문자열 앞에 f 접두사를 붙여 사용할 수 있다.

  ```
  name = '네오 유니버스'
  star = 5
  f'제가 주자로 사용할 말딸은 {name}입니다. 현재 {star}성입니다.'
  ```
<br>&nbsp; f 문자열 포매팅은 위와 같이 변숫값을 생성한 후 그 값을 참조할 수 있다. 또, 표현식 또한 지원하기에 다음과 같이 활용할 수도 있다.

 ```
 star=3
 f'나는 여신상을 몇 더 모으면 네오 유니버스를 {star+1}성으로 개화할 수 있다.'
 ```
<br>&nbsp; 딕셔너리는 f 문자열 포매팅에서 다음과 같이 활용할 수 있다.

```
d={'name':'네오 유니버스', 'star':5}
f'제가 주자로 사용할 말딸은 {d['name']}입니다. 현재 {d['star']} 성입니다.
```


### 문자열 관련 함수들
* 문자 개수 세기 - Count
```
a="umamusme"
a.count('m')
3
```
count 함수를 사용하여 문자열의 m 개수를 리턴했다.
<br>
* 위치 알려 주기 - find, index
```
a="umamusme"
a.find('m')
1 < 문자열에서 m이 처음 나온 위치
a. find ('k')
-1
```
&nbsp; find 함수로 문자열 중 문자 m이 처음으로 나온 위치를 찾았다. 찾는 문자나 문자열이 존재하지 않는다면 -1을 반환한다.


```
a='Life is too short'
a.index('t')
8
a.index('k')
error < 찾는 문자 k가 없으므로 오류 발생
```
&nbsp; index함수로 문자열 중 문자 t가 맨 처음으로 나온 위치를 반환했다. 만약 찾는 문자나 문자열이 존재하지 않는다면 오류가 발생한다. 앞의 find함수와 다른 점은 문자열 안에 존재하지 않는 문자를 찾으면 오류가 발생한다는 점이다.

<br>

* 문자열 삽입 - join

```

a.lstrip()
a= "hi   "
a.rstrip()
a="hi"
```
&nbsp; lstrip과 rstrip은 각각 왼쪽, 오른쪽 방향에 있는 한 칸 이상의 연속된 공백을 모두 지운다. l은 left, r은 right를 의미한다.

* 양쪽 공백 지우기 - strip
```
a="   hi   "
a.strip()
"hi"
```
&nbsp; strip은 문자열 좌우 한 칸 이상의 연속된 공백을 모두 지운다.<br>

* 문자열 바꾸기 - replace
```
a = "You are so pretty"
a.replace("you", "UMAs")
'UMAs are so pretty'
```
&nbsp; replace 함수는 replace(바꿀 문자열,바뀔 문자열)과 같이 사용하여 특정한 값을 치환해 준다. <br>

* 문자열 나누기 - split
```
a='life is too short'
a.split() < 공백을 기준으로 나눔
['life', 'is', 'too', 'short']
b='a:b:c:d'
b.split(':') < :를 기준으로 나눔
['a', 'b', 'c', 'd']
```
&nbsp; split 함수는 인자 입력이 없을 때 공백을 기준으로 문자열을 나누어 준다. 만약 b.split(':')과 같이 괄호 안에 특정 값이 있을 경우 괄호 안의 값을 구분자로 하여 문자열을 나눈다. 이렇게 나눈 문자열은 **리스트에 하나씩 들어간 형태로 저장된다.** <br>

* 기억할 점

위에서 소개한 문자열 처리 함수들은 **변수 자체를 바꾸지 않는다.** a.upper()를 수행하더라도 a.upper은 대문자로 바꾼 값을 출력할 뿐, a 자체를 바꾸지 않음에 주의한다.
